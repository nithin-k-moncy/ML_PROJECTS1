## 1.Logistic Regression

##### split dataset in features and target variablee

```python
x=df.iloc[:,[0,1,2,3,4,5,6]]
y=df.iloc[:,7]
```

##### split x and y into training and testing sets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

##### import the class and fit the model with data

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(solver="lbfgs",max_iter=1000)
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
df2=pd.DataFrame(x_test)
print(df2.to_string())
```

```python
print(y_pred)
```

```python
df2=pd.DataFrame({"actual":y_test,"predicted":y_pred})
print(df2.to_string())
```

##### evaluating the algorithm

```python
accuracy1 = accuracy_score(y_test, y_pred)
print('Accuracy2: %.2f' % (accuracy1*100))
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,
 y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,
 y_pred))
print('Root Mean Squared Error:',
 np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
